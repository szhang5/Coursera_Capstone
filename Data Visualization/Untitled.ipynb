```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
baby_means = [10, 22, 14, 10, 15]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/3, men_means, width, label='Men')
rects2 = ax.bar(x + width/3, women_means, width, label='Women')
rects3 = ax.bar(x + width/3*2, baby_means, width, label='Baby')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width()/3, height),
                    xytext=(0,3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()
fig.set_size_inches(20, 8)

plt.show()
```

```python
df = pd.read_csv('Topic_Survey_Assignment.csv');
df = df.sort_values(by=['Very interested'], ascending=False)
df['Very interested'] = df['Very interested']/2233
df['Somewhat interested'] = df['Somewhat interested']/2233
df['Not interested'] = df['Not interested']/2233
df.head()
```

```python
colorlist = ['#5cb85c', '#5bc0de', '#d9534f']
df.plot(kind='bar', 
        figsize=(20,8), 
        width=0.8, 
        color=colorlist, 
        fontsize=14,
        yticks=[],
       )
plt.title('Percentage of Respondents\'interest in Data Science Areas', fontsize=16)
plt.legend(loc='upper right', fontsize=14)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
positions = [-width/3*2, width/3, width/3*2]

for i in range(df.shape[0]):
    for index, value in enumerate(df):
        label = '{0:.2f}'.format(df[value][i])
        plt.annotate(label,
                    xy=(i+positions[index], df[value][i]),
                    ha='center', 
                    va='bottom',)
        
plt.show()
```
